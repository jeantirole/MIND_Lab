```python
#--- Decoding code completion

cnt = 0 

decoded_time = []

decoded_msgs = []
while cnt < len(df):    

    try:

        # time 
        decoded_time.append( df.iloc[cnt]["day"] + "-" + df.iloc[cnt]["time"][:-5]+"]" )

        # msgs
        row = df.iloc[cnt]["ais_raw"].split(",")
        
        frag_1 = row[1]
        #print("frag_1 " , frag_1)
        frag_2 = row[2]
        
        if frag_1 =='1':
            decoded_msgs.append( decode(df.iloc[cnt]["ais_raw"] ) )
            cnt += 1 
        elif (frag_1 =='2') and (frag_2 =='1'):
            decoded_msgs.append( decode(df.iloc[cnt]["ais_raw"] , df.iloc[cnt+1]["ais_raw"] ) )
            cnt += 2 
        elif (frag_1 =='3') and (frag_2 =='1'):
            decoded_msgs.append( decode(df.iloc[cnt]["ais_raw"] , df.iloc[cnt+1]["ais_raw"],df.iloc[cnt+2]["ais_raw"] ) )
            cnt += 3
            
        if cnt % 1000 ==0:
            print("#---- cnt : ", cnt)
        
        if cnt > len(df):
            break
    
    except Exception as error:
        
        print(df.iloc[cnt]["ais_raw"] )
        decoded_msgs.append(error)
        cnt +=1
```

```python
#--- filter dynamic msgs
dynamic_message_type = [1,3]
dynamic_dict ={
    "mmsi_": [],
    "time" : [],
    "lon_" : [],
    "lat_" : [],
    "heading_" : [],
    "turn_" : [],
    "speed_" : []
}


for i in range(len(W_)):
        
        #--- filter dynamic msgs
        try:
            if W_.iloc[i]["msgs"].msg_type in dynamic_message_type:
                #print(W_.iloc[i]["msgs"])
                
                dynamic_dict["time"].append( pd.to_datetime( W_.iloc[i]['time'][1:-1] ) )

                dynamic_dict["mmsi_"].append(W_.iloc[i]["msgs"].mmsi )
                dynamic_dict["turn_"].append(W_.iloc[i]["msgs"].turn )
                dynamic_dict["speed_"].append(W_.iloc[i]["msgs"].speed )
                dynamic_dict["lon_"].append(W_.iloc[i]["msgs"].lon )
                dynamic_dict["lat_"].append(W_.iloc[i]["msgs"].lat )
                dynamic_dict["heading_"].append(W_.iloc[i]["msgs"].heading )
        except Exception as e:
            print(e)
            pass
```
