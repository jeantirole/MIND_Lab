```python
#--- split json file 
# 소스코드의 tools 폴더에 있는 ms_train.json 파일을 옵션으로 사용 
json_path = "./tools/data/dota/split/split_configs/ms_train.json"

import json 
with open(json_path, "r") as f: 
    js_f = json.load(f)
```

# Valid Preprocessing

- DOTA validation label 안에 있는 데이터 전처리 

```python
# Validation preprocessing 

from glob import glob
import os

# 앞서 설정한 00.Data 폴더 안에, DOTA_dataset 폴더 안에서 작업. 
data_root = "/mnt/hdd/eric/.tmp_ipy/00.Data"
valid_ann = os.path.join(data_root, "DOTA_dataset/dota1_origin/val/annfiles")
validation_ann_path = glob(f"{valid_ann}/*")
```

```python
# 레이블 전처리 코드 

cnt =0
for file_name in validation_ann_path:
    f = open(file_name, 'r')
    lines = f.readlines()
    
    nf = open(f"{file_name}","w")
        
    for line in lines:
        #print(line)      
        if "imagesource" in line:
            pass
        elif "gsd" in line:
            pass
        else:
            nf.write(line)
    nf.close()       
    f.close()

    cnt +=1
    print(cnt)
    #if cnt==1:
    #    break
```
