```python
import os

if os.path.exists(root):
    print("folder exists")
    pass
else:
    os.makedirs(root)
```

```python
image_path = os.path.join(root,"image") 
json_path = os.path.join(root,"json")
```

```python
# verify json to DOTA foramt

json1= os.path.join(json_path,os.listdir(json_path)[0]) 
os.path.exists(json1)
```

```python
from glob import glob 
json_path = json_path+"/*"
json_list = glob(json_path)
```

```python
json_list[0].split("/")[-1][:-5]
```

```python
root
```

```python
import json

# main code 
for json_file in json_list:
    tmp_bl=[]
    with open(json_file,'r') as file:
        da = json.load(file)
        json_name = json_file.split("/")[-1][:-5]

        for pro in da['features']:
            print(pro['properties']['type_name'])
            tmp_bl.append(pro['properties']['type_name'])
```

```python
set(tmp_bl)
```

```python
root
```

```python
# from json to DOTA txt 

import json

# main code 
for json_file in json_list:
    tmp_bl=[]
    with open(json_file,'r') as file:
        da = json.load(file)
        json_name = json_file.split("/")[-1][:-5]

        for pro in da['features']:
            
            # filter class 
            if pro['properties']['type_name'] == 'oil tank': 
                s =pro['properties']['object_imcoords']
                tmp_bl.append([s])

        if not os.path.exists(f"{root}/dota_label_txt"):
            os.mkdir(f"{root}/dota_label_txt")
        f=open(f"{root}/dota_label_txt/{json_name}.txt", "w")

        for a in tmp_bl:
            for box in a:
                box = box.split(",")
                box = [float(i) for i in box]
                print(box[0],box[1],box[2],box[3],box[4],box[5],box[6],box[7], "oiltank",0, "\n")
                f.write(f"{box[0]} {box[1]} {box[2]} {box[3]} {box[4]} {box[5]} {box[6]} {box[7]} oiltank 0 \n")
```
